```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv("SubsetTreatment_response.csv")
file.head()
```

```python
for index, row in file.iterrows():
    if row["DaysFromDxStart"] >= row["DaysFromDxStop"]:
        temp = row["DaysFromDxStart"]
        row["DaysFromDxStart"] = row["DaysFromDxStop"]
        row["DaysFromDxStop"] = temp
```

```python
ids = file["ID"].unique()
print ("The following are the inconsistensies of start and stop days. What do I do with these?")
for ID in ids:
    sub = file[file["ID"] == ID]
    
    line1 = sub[sub["Line"] == 1]
    line2 = sub[sub["Line"] == 2]
    line3 = sub[sub["Line"] == 3]

    if line2.shape[0] > 0: 
        if min(line2["DaysFromDxStart"]) < max(line1["DaysFromDxStop"]):
            print ("1"+ID)
    if line3.shape[0] > 0:
        if max(line2["DaysFromDxStop"]) > min(line3["DaysFromDxStart"]):
            print ("3"+ID)
```

```python
file[(pd.isnull(file["DaysFromDxStop"])) & (file["Line"] == 2)]
print ("No NA stop day for Line 2 has been found.")
```

```python
#after the days errors been fixed:
file["PFS"] = np.nan
for ID in ids:
    sub = file[file["ID"] == ID]

    line2 = sub[sub["Line"] == 2]
    line3 = sub[sub["Line"] == 3]
    
    if line3.shape[0] > 0:
        PFS = min(line3["DaysFromDxStart"]) - min(line2["DaysFromDxStart"])
        file.loc[((file["Line"] == 2) & (file["ID"] == ID)), ["PFS"]] = PFS
```

```python
only_line2 = file[file["Line"] == 2]
```
